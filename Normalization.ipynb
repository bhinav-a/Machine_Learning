```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

since the null value of CoApllicantIncome is 0 we will use coapplicant income 


```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show
```

Outliers doesnt make effect on scaling 
its magnitude gets reduced


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome min"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head()
```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["CoapplicantIncome min"])
plt.show
```
