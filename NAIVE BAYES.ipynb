```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("placement.csv")
dataset.head()
```

```python
sns.scatterplot(x = "cgpa" , y = "iq" , data = dataset , hue="placement")
```

```python
x = dataset.iloc[: , :-1]
```

```python
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
sns.kdeplot(data = dataset["cgpa"])
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train , y_train)
```

```python
gnb.score(x_test , y_test)*100
```

```python
gnb.predict([[6.8 , 123]])
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train , y_train)
```

```python
mnb.score(x_test , y_test)*100
```

```python
gnb.predict([[6.8 , 123]])
```

```python
bnb = BernoulliNB()
```

```python
bnb.fit(x_train ,y_train)
```

```python
bnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf = gnb)
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf = mnb)
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf = bnb)
```
