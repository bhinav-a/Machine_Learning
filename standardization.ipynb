# Standardization :

it is a very effective technique which re scales a features value so that it has distribution with 0 mean value and variance eqauls to 1


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean() , inplace = True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import  StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]) , columns = ["x"])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["ApplicantIncome_ss"])

plt.show()
```
